```python
import os
import numpy as np
import shutil
```

```python
N_images = 300
ad_src  = "/data/csc4801/WangLab/images_organizedAD/"
mci_src = "/data/csc4801/WangLab/images_organizedMCI/"
nl_src  = "/data/csc4801/WangLab/images_organizedNL/"
ad_dest = "/data/csc4801/WangLab/training/AD/"
mci_dest = "/data/csc4801/WangLab/training/MCI/"
nl_dest = "/data/csc4801/WangLab/training/NL/"
```

```python
ad_train = np.random.choice(os.listdir(ad_src), size=N_images // 3, replace=False)
mci_train = np.random.choice(os.listdir(mci_src), size=N_images // 3, replace=False)
nl_train = np.random.choice(os.listdir(nl_src), size=N_images // 3, replace=False)
```

```python
os.makedirs(ad_dest, exist_ok=True)
os.makedirs(mci_dest, exist_ok=True)
os.makedirs(nl_dest, exist_ok=True)
```

```python
added = []
print(os.listdir(ad_dest))
for ad in ad_train:
    patient_dir = os.path.join(ad_src, ad)
    print(ad)
    if not ad in os.listdir(ad_dest):
        while ad in ad_train \
              or ad in added \
              or len(os.listdir(patient_dir)) == 0:
            ad = np.random.choice(os.listdir(ad_src), replace=False)
            patient_dir = os.path.join(ad_src, ad)
        if not ad in os.listdir(ad_dest):
            shutil.copytree(patient_dir, os.path.join(ad_dest, ad))
            added.append(ad)
```

```python
added = []
for mci in mci_train:
    patient_dir = os.path.join(mci_src, mci)
    if not mci in os.listdir(mci_dest):
        while mci in mci_train \
              or mci in added \
              or len(os.listdir(patient_dir)) == 0:
            mci = np.random.choice(os.listdir(mci_src), replace=False)
            patient_dir = os.path.join(mci_src, mci)
        if not mci in os.listdir(mci_dest):
            shutil.copytree(patient_dir, os.path.join(mci_dest, mci))
            added.append(mci)
```

```python
added = []
for nl in nl_train:
    patient_dir = os.path.join(nl_src, nl)
    if not nl in os.listdir(nl_dest):
        while nl in nl_train \
              or nl in added \
              or len(os.listdir(patient_dir)) == 0:
            nl = np.random.choice(os.listdir(nl_src), replace=False)
            patient_dir = os.path.join(nl_src, nl)
        if not nl in os.listdir(nl_dest):
            shutil.copytree(patient_dir, os.path.join(nl_dest, nl))
            added.append(nl)
```
